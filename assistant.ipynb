```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Setup Chrome driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

def send_email_gmail(username, password, to, subject, body):
    # Open Gmail login page
    driver.get("https://mail.google.com/")
    
    # Log into Gmail
    email_box = driver.find_element(By.ID, "identifierId")
    email_box.send_keys(username)
    email_box.send_keys(Keys.RETURN)
    time.sleep(2)

    # Enter password
    password_box = driver.find_element(By.NAME, "password")
    password_box.send_keys(password)
    password_box.send_keys(Keys.RETURN)
    time.sleep(5)
    
    # Click on Compose button to open the new message window
    compose_button = driver.find_element(By.XPATH, "//div[contains(text(), 'Compose')]")
    compose_button.click()
    time.sleep(2)
    
    # Fill in the recipient email
    to_box = driver.find_element(By.NAME, "to")
    to_box.send_keys(to)
    
    # Fill in the subject
    subject_box = driver.find_element(By.NAME, "subjectbox")
    subject_box.send_keys(subject)
    
    # Fill in the body of the email
    body_box = driver.find_element(By.CSS_SELECTOR, "div[aria-label='Message Body']")
    body_box.send_keys(body)
    
    # Click on Send button
    send_button = driver.find_element(By.XPATH, "//div[contains(text(), 'Send')]")
    send_button.click()
    time.sleep(3)
    
    print("Email sent successfully!")
    
    # Close the browser
    driver.quit()

# Usage
username = "your_email@gmail.com"  # Your Gmail email address
password = "your_gmail_password"  # Your Gmail password (or App Password if 2FA is enabled)
to = "recipient_email@example.com"  # Recipient's email address
subject = "Test Email from Selenium"
body = "Hello, this is a test email sent from Selenium using Gmail."

send_email_gmail(username, password, to, subject, body)
```

```python
import pyttsx3
import datetime
import time
import os
import webbrowser
import pyjokes
import wikipedia
import smtplib
import speech_recognition as sr
import sys
import psutil
# import pyautogui
import json
import numpy as np
import pickle
import time
from pywhatkit import sendwhatmsg
from nltk.chat.util import Chat
from keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
import random

# Initialize Text-to-Speech engine
def initialize_engine():
    engine = pyttsx3.init("sapi5")
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[1].id)
    rate = engine.getProperty('rate')
    engine.setProperty('rate', rate - 50)
    volume = engine.getProperty('volume')
    engine.setProperty('volume', volume + 0.25)
    return engine

# Speak function
def speak(text):
    engine = initialize_engine()
    engine.say(text)
    engine.runAndWait()

# Greet the user based on the time of day
def wish_user():
    hour = int(datetime.datetime.now().hour)
    if hour >= 0 and hour < 12:
        speak("Good Morning!")
    elif hour >= 12 and hour < 16:
        speak("Good Afternoon!")
    else:
        speak("Good Evening!")
    speak("I am your Desktop Assistant. How may I assist you today?")

# Command listener
def user_command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("I am listening....")
        r.pause_threshold = 2
        audio = r.listen(source)

    try:
        print("Recognizing.....")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}")
    except Exception as e:
        print("Sorry, I didn't catch that. Please say again...")
        return "None"
    return query

# Schedule for the day
def schedule():
    day = datetime.datetime.today().weekday() + 1
    day_dict = {
        1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday", 7: "Sunday"
    }
    day_of_week = day_dict.get(day, "Invalid day")
    week_schedule = {
        "monday": "From 9:00 to 9:50 you have Algorithms class...",
        "tuesday": "From 9:00 to 9:50 you have Web Development class...",
        "wednesday": "Full day of classes, including Machine Learning and Operating Systems...",
        "thursday": "Full day of classes, including Computer Networks and Cloud Computing...",
        "friday": "Full day of classes, including AI and Advanced Programming...",
        "saturday": "Team meetings and personal development time...",
        "sunday": "Holiday, but catch up on any pending tasks."
    }
    speak(f"Today is {day_of_week}. Here's your schedule: {week_schedule.get(day_of_week.lower(), 'No schedule found for today.')}")

# Social media opening function
def social_media(command):
    if 'facebook' in command:
        speak("Opening Facebook...")
        webbrowser.open("https://www.facebook.com/")
    elif 'whatsapp' in command:
        speak("Opening WhatsApp...")
        webbrowser.open("https://web.whatsapp.com/")
    elif 'instagram' in command:
        speak("Opening Instagram...")
        webbrowser.open("https://www.instagram.com/")
    elif 'discord' in command:
        speak("Opening Discord...")
        webbrowser.open("https://discord.com/")
    else:
        speak("Sorry, I couldn't find that application.")

# Play music or video
def play_music(query):
    if "spotify" in query:
        speak("Opening Spotify...")
        webbrowser.open("https://www.spotify.com")
    elif "youtube" in query:
        speak("Opening YouTube...")
        webbrowser.open("https://www.youtube.com")

# Send an email function
def mail_sent(to, content):
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.ehlo()
    server.starttls()
    server.login('your-email@example.com', 'your-app-password')  # Use app password
    server.sendmail('your-email@example.com', to, content)
    server.close()

# Joke function
def tell_joke():
    joke = pyjokes.get_joke()
    speak(joke)

# CPU and battery status check
def system_condition():
    cpu_usage = psutil.cpu_percent()
    speak(f"CPU usage is at {cpu_usage}%")
    battery = psutil.sensors_battery()
    battery_percentage = battery.percent
    speak(f"Battery is at {battery_percentage}%")
    if battery_percentage >= 80:
        speak("Battery is sufficiently charged.")
    elif battery_percentage >= 40 and battery_percentage <= 75:
        speak("Battery is moderate, consider charging.")
    else:
        speak("Battery is low, please connect to a charger.")

# Main function to run the assistant
if _name_ == "_main_":
    wish_user()

    while True:
        query = user_command().lower()

        if 'wikipedia' in query:
            speak("Searching Wikipedia...")
            query = query.replace("wikipedia", "")
            summary = wikipedia.summary(query, sentences=3)
            speak("Here is what I found: " + summary)

        elif 'open youtube' in query:
            play_music(query)

        elif 'search google' in query:
            speak("What should I search on Google?")
            query = user_command().lower()
            webbrowser.open(f"https://www.google.com/search?q={query}")

        elif 'send email' in query:
            speak("Please tell me the content of the email.")
            content = user_command()
            speak("Whom should I send the email to?")
            to = input("Enter recipient email: ")
            mail_sent(to, content)
            speak(f"Email sent to {to}")

        elif 'tell me a joke' in query:
            tell_joke()

        elif 'open facebook' in query or 'open instagram' in query or 'open whatsapp' in query or 'open discord' in query:
            social_media(query)

        elif 'what is the time' in query:
            current_time = datetime.datetime.now().strftime("%H:%M")
            speak(f"The time is {current_time}.")

        elif 'date today' in query:
            today = datetime.datetime.today().strftime("%B %d, %Y")
            speak(f"Today is {today}.")

        elif 'open calculator' in query:
            os.startfile("C:\\Windows\\System32\\calc.exe")
            speak("Opening Calculator.")

        elif 'open notepad' in query:
            os.startfile("C:\\Windows\\System32\\notepad.exe")
            speak("Opening Notepad.")

        elif 'close calculator' in query:
            os.system("taskkill /f /im calc.exe")
            speak("Closing Calculator.")

        elif 'close notepad' in query:
            os.system("taskkill /f /im notepad.exe")
            speak("Closing Notepad.")

        elif 'system condition' in query or 'battery status' in query:
            system_condition()

        elif 'exit' in query or 'quit' in query:
            speak("Goodbye! Have a great day.")
            sys.exit()

        time.sleep(2)  # Wait for the next command
```
